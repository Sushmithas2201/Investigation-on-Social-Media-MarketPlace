## Author Information
authorMeta/fans: Number of fans or followers

authorMeta/following: Number of accounts the author is following

authorMeta/friends: Number of friends

authorMeta/heart: Total number of hearts/likes received

authorMeta/id: Unique identifier for the author

authorMeta/name: Author's username

authorMeta/nickName: Author's display name

authorMeta/privateAccount: Indicates if the account is private

authorMeta/signature: Author's bio or profile signature

authorMeta/ttSeller: Indicates if the author is a TikTok seller

## Video Information
commentCount: Number of comments on the video

diggCount: Number of "diggs" or likes

playCount: Number of times the video has been played

shareCount: Number of times the video has been shared

submittedVideoUrl: URL of the submitted video

text: Text or caption of the video

videoMeta/definition: Video quality (e.g., 540p)

videoMeta/duration: Duration of the video in seconds

videoMeta/format: Video format (e.g., mp4)

videoMeta/height: Height of the video

videoMeta/width: Width of the video

webVideoUrl: URL to view the video on the web

### Correlation Analysis Between different Metrics ###

### Hastag Analysis ###

```python
hashtag_columns =[col for col in data.columns if 'hashtags/' in col]
hashtags_data =data[hashtag_columns]

#add it into single list
hashtags = hashtags_data.values.flatten()
hashtags=hashtags[~pd.isnull(hashtags)]

#frequency of each hashtags
hashtag_counts=pd.Series(hashtags).value_counts()
top_hashtags=hashtag_counts.head(10)

#Top Hashtags
plt.figure(figsize=(10, 6))
sns.barplot(x=top_hashtags.index, y=top_hashtags.values, color="dodgerblue")
plt.title('Top Hashtags Used in Videos')
plt.xlabel('Hashtag')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
display(top_hashtags)
```

### Text Content Analysis ###

```python
#This is to identify common themes and keywords
from sklearn.feature_extraction.text import CountVectorizer

#Data extraction and cleaning the text data
text_data=data['text'].dropna().str.lower()

#Use of CounterVector to find the most common words
vectorizer = CountVectorizer(stop_words='english', max_features=10)
word_counts= vectorizer.fit_transform(text_data)

#Get the top words and their counts
word_count_df =pd.DataFrame(word_counts.toarray(), columns=vectorizer.get_feature_names_out())
top_words = word_count_df.sum().sort_values(ascending=False)

display(top_words)

```

### Video Duration Analysis ###

## Most frequently used hashtags across all Regions

```python
# initialize a dictionary to store hastags by location
regional_hashtag_dict = {}

# iterate through each row to collect hashtags by locationn
for _, row in data.iterrows():
    location = row['locationCreated']
    hashtags = row[hashtag_columns].dropna().tolist()
    
    #Filter out the numerical hashtags
    hashtags = [str(tag) for tag in hashtags if not str(tag).isnumeric()]
    
    if location not in regional_hashtag_dict:
        regional_hashtag_dict[location]=[]
        
    regional_hashtag_dict[location].extend(hashtags)
#convert the dictionary to DataFrame
regional_hashtag_df= pd.DataFrame([
    {'location':loc, 'Hashtag': hashtag}
    for loc, hashtags in regional_hashtag_dict.items()
    for hashtag in hashtags
])

#frequency of each hashtag by location
regional_hashtag_counts = regional_hashtag_df['Hashtag'].value_counts().reset_index(name='Count')
regional_hashtag_counts.columns = ['Hashtag', 'Count']

#top regional hashtags
top_regional_hashtags = regional_hashtag_counts.head(10)



top_regional_hashtags
```

## Upload Timing Analyis ##

##### Reference: https://dfir.blog/tinkering-with-tiktok-timestamps/

```python
#convert date from unix date time stamp
data['createTime']=pd.to_datetime(data['createTime'],unit='s')
data['createDate']=data['createTime'].dt.date
data['upload_hour']=data['createTime'].dt.hour
data['upload_day']=data['createTime'].dt.day_name()


hourly_enagement=data.groupby('upload_hour')[['commentCount','diggCount','playCount','shareCount']].mean()


plt.figure(figsize=(14,6))
hourly_enagement.plot(marker='o')
plt.title('Average Enagagement by Upload Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Average Enagagement')
plt.legend(title='Engagement Metrics')
plt.grid(True)
plt.xticks(range(0,24))
plt.show()
```
